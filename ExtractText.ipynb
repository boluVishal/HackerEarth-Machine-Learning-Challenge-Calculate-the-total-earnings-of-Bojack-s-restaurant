```python
import os, io
from google.cloud import vision
import pandas as pd
entries = os.listdir('./Test Image/')
iname = []
for entry in entries:
    iname.append(entry) 
#enter your cred file here
os.environ['GOOGLE_APPLICATION_CREDENTIALS'] = r'XYZ.json'
client = vision.ImageAnnotatorClient()

for i in iname:

        file_name = i
        image_path = f'Test Image/'+i

        with io.open(image_path, 'rb') as image_file:
            content = image_file.read()

        # construct an iamge instance
        image = vision.types.Image(content=content)


        # annotate Image Response
        response = client.text_detection(image=image)  # returns TextAnnotation
        df = pd.DataFrame(columns=['locale', 'description'])

        texts = response.text_annotations
        #print(texts)
        for text in texts:
            df = df.append(
                dict(
                    locale=text.locale,
                    description=text.description
                ),
                ignore_index=True
            )

        #print(df['description'][0])
        print(i)
        with open('./Extracted text/'+i+'.txt',mode ='w',encoding="utf-8") as file:      

                         file.write(df['description'][0]) 

```
